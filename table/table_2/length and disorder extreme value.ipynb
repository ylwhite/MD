## Disorder avg and var

```python
disorder_R_H = pickle.load(open('/home/dldx/R-H/code/classification/ML/data/disorder/avg_R-H_all_dsiorder.pkl','rb'))
```

```python
t = np.log(np.array([i[9] for i in disorder_R_H[1]]))
new_arr = np.delete(t , 84)
print(t)
print(new_arr)
```

```python
# 补充IDR10出现异常值的情况
replace = 0
t = np.log(np.array([i[9] for i in disorder_R_H[1]]))
new_t = np.delete(t , 84)
print("期望值：" + str(np.mean(new_t)))
print("方差：" + str(np.var(new_t)))
```

```python
# 补充IDR10出现异常值的情况
replace = 0
t = np.log(np.array([i[9] for i in disorder_R_H[1]]))
t[84] = replace
print("期望值：" + str(np.mean(t)))
print("方差：" + str(np.var(t)))
```

```python
for k in range(21): 
    R_one = np.log(np.array([i[k] for i in disorder_R_H[0]]))
    H_one = np.log(np.array([i[k] for i in disorder_R_H[1]]))
    print("期望值：" + str(np.mean(R_one)))
    print("期望值：" + str(np.mean(H_one)))
    print("方差：" + str(np.var(R_one)))
    print("方差：" + str(np.var(H_one)))    
    print("==============IDR" + str(k+1) + "======================")
```
