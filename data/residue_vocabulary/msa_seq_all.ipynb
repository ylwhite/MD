```python
%pylab inline
import pandas as pd
import numpy as np
import graphviz
import pysam as sam
```

```python
fasta = sam.FastaFile("/home/dldx/R-H/code/classification/ML/data/mafft_MSA_R-H_out.fa")
```

```python
fasta.references
```

```python
len(fasta.references)
```

```python
name = []
seq = []
tier = []
for i in range(72):
    name.append(fasta.references[i])
    seq.append(fasta.fetch(fasta.references[i],0,fasta.lengths[i]))
    tier.append(1)
for j in range(72,179):
    name.append(fasta.references[j])
    seq.append(fasta.fetch(fasta.references[j],0,fasta.lengths[j]))
    tier.append(0)
```

```python
data = [name,seq,tier]
# 用字典设置DataFrame所需数据
dfData = { 
        'name':data[0],
        'Sequence':data[1],
        'Tier':data[2],
    }
df = pd.DataFrame(dfData) # 创建DataFrame

df.to_excel('/home/dldx/R-H/code/classification/ML/data//mafft_MSA_R_H.xlsx',index=False) # 存表，
```
