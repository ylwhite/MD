## 2、将获取的disorder值对应赋予给做完MSA维度统一序列（-赋值为0）

```python
#读取 pickle文件
R_1 = pickle.load(open('/home/dldx/R-H/code/classification/ML/data/disorder/all_seq_disorder_R_1.pkl','rb'))
H_3 = pickle.load(open('/home/dldx/R-H/code/classification/ML/data/disorder/all_seq_disorder_H_3.pkl','rb'))
```

```python
R_1_seq_disorder = []
for i in range(72):
    k = 0
    one_seq_disorder = []
    msa_seq = fasta.fetch(fasta.references[i],0,fasta.lengths[i])
    for j in range(len(msa_seq)):
        if msa_seq[j] == "-":
            one_seq_disorder.append(0)
        else:
            one_seq_disorder.append(R_1[i][k])
            k = k + 1
    R_1_seq_disorder.append(one_seq_disorder)

H_3_seq_disorder = []
for i in range(72,179):
    k = 0
    one_seq_disorder = []
    msa_seq = fasta.fetch(fasta.references[i],0,fasta.lengths[i])
    for j in range(len(msa_seq)):
        if msa_seq[j] == "-":
            one_seq_disorder.append(0)
        else:
            one_seq_disorder.append(H_3[i-72][k])
            k = k + 1
    H_3_seq_disorder.append(one_seq_disorder)
```

```python
# 制定表格
seq_name = []
all_R_H_disorder = []
Tier = []
seq = []
for i in range(72):
    seq_name.append(fasta.references[i])
    all_R_H_disorder.append(R_1_seq_disorder[i])
    Tier.append(1)
    seq.append(fasta.fetch(fasta.references[i],0,fasta.lengths[i]))
for j in range(107):
    seq_name.append(fasta.references[j+72])
    all_R_H_disorder.append(H_3_seq_disorder[j])
    Tier.append(0) 
    seq.append(fasta.fetch(fasta.references[j+72],0,fasta.lengths[j+72]))
    
data = [seq_name,all_R_H_disorder,Tier,seq]
# 用字典设置DataFrame所需数据
dfData = { 
        'seq_name':data[0],
        'seq_all_disorder':data[1],
        'Tier':data[2],
        'Sequence':data[3],
    }
df = pd.DataFrame(dfData) # 创建DataFrame

df.to_csv('/home/dldx/R-H/code/classification/ML/data/disorder/MSA_disorder_R_H.csv',index=False) # 存表，
```
