```python
all_21_IDR_disorder = []
for position in range(21):
    one_IDR = []
    R_IDR = [i[position] for i in disorder_R_H[0]]
    H_IDR = [i[position] for i in disorder_R_H[1]]
 
    # 转化为二维数组
    for i in range(72):
        test =[]
        test.append(R_IDR[i])
        one_IDR.append(test)
    for i in range(107):
        test =[]
        test.append(H_IDR[i])
        one_IDR.append(test)

    # 所有IDR的平均disorder值添加到一个数组
    all_21_IDR_disorder.append(one_IDR)
```
