```python
def alntxt(aln_path):
    fr=open(aln_path, 'r')  #rmsd tiedla,2条序列
    dic={}
    for line in fr:
        if line.startswith('>'):
            name=line.split()[0]
            name=name[1:15]
            dic[name]=''
        else:
            dic[name]+=line.replace('\n','')#构建字典
    fr.close()
    list_tied=[]
    list_tied_name=[]
    ss=sorted(dic.items())
    for k,v in ss:
        list_tied_name.append(k)
        list_tied.append(v)
    print(list_tied_name)
    return list_tied_name,list_tied  
```
