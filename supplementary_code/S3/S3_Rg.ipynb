```python
# R_open
colors = iter(cm.Reds(np.linspace(0.39, 1, 10)))
figure(figsize=(10,6))
for i in range(1,11):
    one_top_h = md.load("/www/yang/R_open/r2_bound.pdb")
    one_traj = md.load("/www/yang/R_open/r" +str(i) + "/pro.xtc",  top=one_top_h)
    one_rg = md.compute_rg(one_traj)
    plot(one_rg, color = next(colors),linewidth=0.36)
font={'size':15,
     'weight':'bold',}
legend(["r1","r2","r3","r4","r5",
        "r6","r7","r8","r9","r10"],
        loc="upper center",ncol=5,prop=font)
plt.text(100,3.24, '%s'%('A'), weight='bold',fontsize=15)
xlim(0,10100)
xticks(range(0,10100,1000),range(0,101,10),weight='bold',fontsize=15)
xlabel('Time (ns)', weight='bold',fontsize=16)
ylim(2.3,3.3)
yticks([2.4,2.6,2.8,3.0,3.2],weight='bold',fontsize=15)
ylabel('Rg (nm)',weight='bold',fontsize=16)
plt.savefig("/home/dldx/R-H/code/supplementary/S3/R_open_rg.jpg",dpi=900)
```
